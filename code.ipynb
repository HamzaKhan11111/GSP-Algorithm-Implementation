# extract_frequent_sequences function which takes list of sequences and support count as input and returns frequent sequences
### Also print Frequent as well as infrequent sequences

```python
def extract_frequent_sequences(data,sup,k):
    if k==1:
        items=df["ProdDescr"].unique().tolist()
        infreqItems=[]
        freqItems=[]
        for item in items:
            count=0
            for index,row in trans_df.iterrows():
                present=0
                flat_list=list(itertools.chain.from_iterable(row["ProdDescr"]))
                if item in flat_list:
                    count=count+1
            if count>=sup:
                freqItems.append(item)
            else:
                infreqItems.append(item)
        print("Frequent items for k==1 are:")
        print(freqItems)
        print("Infrequent items for k==1 are:")
        print(infreqItems)
        return freqItems
    else:
        infreqItems=[]
        freqItems=[]
        for item in data:
            #single event, k item
            if type(item)==type((1,2)):
                count=0
                for index,row in trans_df.iterrows():
                    present=0
                    flat_list=list(itertools.chain.from_iterable(row["ProdDescr"]))
                    for lis in row["ProdDescr"]:
                        if len(set(item).intersection(set(lis)))==len(item):
                            present=1
                            break
                    if present==1:
                        count+=1
            #k event single item
            else:
                count=0
                for index,row in trans_df.iterrows():
                    itemIndex=0
                    for ele in row["ProdDescr"]:
                        if item[itemIndex] in ele:
                            itemIndex+=1
                        if itemIndex==len(item):
                            count+=1
                            break
            if count>=sup:
                freqItems.append(item)
            else:
                infreqItems.append(item)
        print("Frequent items for k==",k," are:")
        print(freqItems)
        print("Infrequent items for k==",k," are:")
        print(infreqItems)
        return freqItems
```

# makeCombinations function makes possible combinations of length k+1 from previous frequent sequences of length k
### (Note: It does not make all possible combinations but only combinations from previous sequences)

```python
def makeCombinatations(data,k):
    if k == 1:
        lis_a=list(itertools.combinations(data,k+1))

        comb = list(itertools.combinations_with_replacement(data,k+1))
        lis_b=[list(combination) for combination in comb]
        
        return lis_a+lis_b
    else:
        elements=[]
        for lis in data:
            for lis2 in lis:
                if lis2 not in elements:
                    elements.append(lis2)
        print(elements)
        final=[]

        lis_a=list(itertools.combinations(elements,k+1))

        if len(lis_a)!=0:
            final.append(lis_a[0])

        new_list = [x for x in data if not isinstance(x,tuple)]
        newlis=[]
        for i in new_list:
            for j in elements:
                possibility1=i.copy()
                possibility1.append(j)

                possibility2=i.copy()
                possibility2.insert(0,j)

                newlis.append(possibility1)
                newlis.append(possibility2)

        for x in newlis:
            if x not in final:
                final.append(x)
        return final
```

# GSP Algorithm function

```python
def GSP(sup):
    k=1
    candidates=[]
    freqItems=[]
    while True:
        if k==5:
            return freqItems
        
        tempItems=extract_frequent_sequences(candidates,sup,k)
        if len(tempItems)!=0:
            freqItems=tempItems
        else:
            return freqItems
        
        candidates=makeCombinatations(freqItems,k)
        if len(candidates)==0:
            return freqItems
        k+=1
```

# Calling GSP function for support of 300

```python
start_time=time.time()
GSP(300)
end_time=time.time()
```
